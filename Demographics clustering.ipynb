```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
#  Import and read the county_data.csv.
county_df = pd.read_csv("csv_files/clustering.csv")
county_df.head()
```

```python
# Initializing model with K = 3 (since we already know there are three classes of iris plants)
model = KMeans(n_clusters=6, random_state=5)
model
```

```python
# Fitting model
model.fit(county_df)
```

```python
# Get the predictions
predictions = model.predict(county_df)
print(predictions)
```

```python
# Add a new class column to the df_iris
county_df["class"] = model.labels_
county_df.head()
```

```python
import plotly.express as px
import hvplot.pandas
```

```python
# Create a scatterplot of df_iris
county_df.hvplot.scatter(x="White", y="deaths", by="class")
```

```python
# Plotting the clusters with three features
fig = px.scatter_3d(county_df, x="White", y="cases", z="deaths", color="class", symbol="class", size="White",width=800)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```

```python
county_df= county_df[["cases", "deaths", "White", "class"]]
county_df.head(10)
```

```python
# Initial imports
import pandas as pd
from sklearn.cluster import KMeans
import plotly.express as px
import hvplot.pandas
```

```python
inertia = []
k = list(range(1, 11))
```

```python
# Calculate the inertia for the range of K values
for i in k:
   km = KMeans(n_clusters=i, random_state=0)
   km.fit(county_df)
   inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```
