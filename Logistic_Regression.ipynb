```python
# Initial imports
import pandas as pd
import numpy as np
#  Import and read the county_data.csv.
cleaned_county_df = pd.read_csv("county_data_cleaned.csv")
cleaned_county_df.head()
```

```python
# Seperate X features from Y features
y = cleaned_county_df["total_votes20"]
X = cleaned_county_df.drop(columns=["total_votes20", "county", "state"])
```

```python
# Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                   )
X_train.shape
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=1000,
                                random_state=1)
```

```python
#Fit (train) or model using the training data
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
