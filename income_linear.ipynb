```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('./csv_files/county_data_cleaned.csv'))
df.head()
```

```python
plt.scatter(df.percentage16_Donald_Trump, df.Income)
plt.xlabel('2016 % of Votes for Donald')
plt.ylabel('Income')
plt.show()
```

```python
plt.scatter(df.percentage20_Donald_Trump, df.Income)
plt.xlabel('2020 % Votes for Donald')
plt.ylabel('Income')
plt.show()
```

```python
plt.scatter(df.percentage16_Hillary_Clinton, df.Income)
plt.xlabel('2016 % Votes for Hillary')
plt.ylabel('Income')
plt.show()
```

```python
plt.scatter(df.percentage20_Joe_Biden, df.Income)
plt.xlabel('2020 % Votes for Joe')
plt.ylabel('Income')
plt.show()
```

```python
X = df.percentage16_Donald_Trump.values.reshape(-1, 1)
```

```python
y = df.Income
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
