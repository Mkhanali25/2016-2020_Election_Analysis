```python
# Import Dependencies
import pandas as pd
from path import Path
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import tree
from sklearn.metrics import accuracy_score
# Create file path for wealth_distribution csv
file_path = Path("csv_files/county_data_cleaned.csv")
county_df = pd.read_csv(file_path)
county_df.head()
```

```python
# Make a copy of the county_df
df2 = county_df.copy()
```

```python
# Drop the state, county, votes20_Donald_Trump, votes20_Joe_Biden, Total_votes20, percentage20_Donald_Trump,
# percentage20_Joe_Biden columns
df2 = df2.drop(["county", "state", "state_num","percentage16_Donald_Trump", "percentage16_Hillary_Clinton",
                "votes20_Donald_Trump", "votes20_Joe_Biden", "total_votes20","percentage20_Donald_Trump",
                "percentage20_Joe_Biden"], axis=1)
df2.head()
```

```python
# Create a function to determine who won each state
def my_fun (df2):
    if (df2["votes16_Donald_Trump"] > df2["votes16_Hillary_Clinton"]):
        return 0
    else:
        return 1
# Create a "who_won" column to display our data
df2['who_won'] = df2.apply(my_fun, axis = 1)
df2.head(10)
```

```python
# Creating the scaler instance
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
# Fitting the scaler
data_scaled = data_scaler.fit_transform(df2)
data_scaled[:5]
```

```python
import numpy as np
print(np.mean(data_scaled[:,0]))
print(np.std(data_scaled[:,0]))
```

```python
# Separate the Features (X) from the Target (y)
y = df2["who_won"]
X = df2.drop(columns="who_won")
```

```python
# Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=900,
                                random_state=1)
```

```python
# Fit (train) or model using the training data¶
classifier.fit(X_train, y_train)
```

```python
# Make predicitions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, y_pred)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, y_pred))
```
